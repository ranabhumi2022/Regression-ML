```python
#----1---
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

data=pd.read_csv("/content/Salary Data.csv")

# Fill missing values in 'Years of Experience' with the mean
data['Years of Experience'].fillna(data['Years of Experience'].mean(), inplace=True)
# Fill missing values in 'Salary' with the mean
data['Salary'].fillna(data['Salary'].mean(), inplace=True)

print(data.head())

x=data['Years of Experience']
y=data['Salary']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

# Reshape x_train and x_test for LinearRegression
x_train = x_train.values.reshape(-1, 1)
x_test = x_test.values.reshape(-1, 1)

model=LinearRegression()
model.fit(x_train,y_train)
prediction=model.predict(x_test)
print(prediction)

accuracy=r2_score(y_test,prediction)
print("R-squared",accuracy)
```
